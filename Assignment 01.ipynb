```python
#Task 01
upper_case=0
lower_case=0
user_input=input()
for char in user_input:
    if 65<=ord(char)<=90:
        upper_case+=1
    elif 91<=ord(char)<=122:
        lower_case+=1
if upper_case>lower_case:
    print(user_input.upper())
else:
    print(user_input.lower())
```

```python
#Task 02
counter1=0
counter2=0
user_input=input()
for item in user_input:
    if 48<=ord(item)<=57:
        counter1+=1
    elif 65<= ord(item)<=90 or 91<=ord(item)<=122:
        counter2+=1
if counter1==len(user_input):
    print('NUMBER')
elif counter2==len(user_input):
    print('WORD')
else:
    print('MIXED')
```

```python
#Task 06
lowcase=0
uppercase=0
digit=0
special=0
user_input=input()
for char in user_input:
    if 65<= ord(char)<=90:
        uppercase+=1
    if 97<= ord(char)<=122:
        lowcase+=1
    if 48<= ord(char)<=57:
        digit+=1
    if  32 <= ord(char)<=47 or 58<= ord(char)<=64 or 91<= ord(char)<=96:
        special+=1
if lowcase!=0 and uppercase !=0 and digit!=0 and special!=0:
    print('OK')
else:
    if uppercase==0:
        print('Uppercase is missing',end=',')
    if digit==0:
        print('Digit is missing',end=',')
    if lowcase==0:
        print('Lower case is missing',end=',')
    if special==0:
        print('Sprcial is missing',end='')
```

```python
#List
#Task 04
l=[]
while True:
    user_input=input()
    if user_input!='STOP':
        l=user_input.split()
    else:
        break
    numeric=[]
    diffrence=[]
    for i in range(int(len(l))-1):
        numeric.append(i+1)
        diff=abs(int(l[i])-int(l[i+1]))
        diffrence.append(diff)
    diffrence.sort()
    if numeric == diffrence:
        print('UB Jumper')
    else:
        print('Not UB jumper')
```

```python
#List
#Task 05
user_input=input()
low=[]
upper=[]
even=[]
odd=[]
for char in user_input:
    if 65<= ord(char)<=90:
        upper.append(char)
        upper.sort()
    elif 97<= ord(char)<=122:
        low.append(char)
        low.sort()
    elif 48<= ord(char)<=57 and int(char)%2==0:
        even.append(char)
        even.sort()
    elif 48<= ord(char)<=57 and int(char)%2!=0:
        odd.append(char)
        odd.sort()
all=low+upper+odd+even
new_str=''
for item in all:
    new_str+=item
print(new_str)
```

```python
#Dictonary & Tuple
#Task 01
user_input=input().split(',')
dict1={}
for i in user_input:
    a=i.split(':')
    dict1[a[0].strip()]=int(a[1])
dict2 = {}
user_input1=input().split(',')
for i in user_input1:
    a=i.split(':')
    dict2[a[0].strip()]=int(a[1])
dict3={}
for k1 in dict1.keys():
    dict3[k1]=0
for k2 in dict2.keys():
    dict3[k2]=0
for k1,v1 in dict1.items():
    dict3[k1]=dict3.get(k1)+v1
for k2,v2 in dict2.items():
    dict3[k2]=dict3.get(k2)+v2
print(dict3)
l=[int(x) for x in dict3.values()]
sorted_tuple=sorted(set(l))
print(tuple(sorted_tuple))
```

```python
#Dictonary & Tuple
#Task 03
string = input()
list1 = list(string.split())
dict1 = {}
list2 = []
list3 = []
for x in range(0,len(list1)):
    if x%3 == 0:
        list2.append(list1[x])
        list3.append(list1[x+2])
list4=[]
for i in list3:
    if i.endswith(','):
        new_Str = i.replace(',', '')
        list4.append(new_Str)
    else:
        list4.append(i)
for x in range(0,len(list4)):
    new=[]
    if list4[x] not in dict1.keys():

        dict1[list4[x]]=list2[x]
    elif list4[x] in dict1.keys():
        for key,value in dict1.items():
            if key==list4[x]:
                new.append(value)
                new.append(list2[x])
                dict1[key]=new
new_l=[]
for key , value in dict1.items():
    z=type(value)
    if z is str:
        new_l.append(value)
        dict1[key]=new_l
print(dict1)
```
